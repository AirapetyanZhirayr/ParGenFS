```python
import requests 
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
url = 'https://link.springer.com/'
```

```python
r = requests.get(url+'journal/357/volumes-and-issues')
soup = BeautifulSoup(r.text, 'html.parser')
results = soup('li', class_='c-list-group__item')
i=0
for result in results:
    issue_link = result.find('a')['href']
#     issue_link = result.a['href']
    r = requests.get(url+issue_link)
    soup = BeautifulSoup(r.text, 'html.parser')
    publications = soup.find_all('div', attrs={'class':'toc-item no-access', 'class':'toc-item'})
    
    for publication in publications:
#         if publication.find('p', 'content-type').text == 'OriginalPaper':
#         if publication.find('p', 'content-type').text in ['OriginalPaper', 'Authors Of Articles', 'Article']:

        paper_link = publication.find('a')['href']
        r = requests.get(url+paper_link)
        paper = BeautifulSoup(r.text, 'html.parser')
        description=paper.find('h2', attrs={'class':
                        'c-article-section__title u-h2 js-section-title js-c-reading-companion-sections-item'})
        if description:
            if 'Abstract' in description.text:

                abstract = paper.find('div', attrs={'class':'c-article-section__content', 'id':'Abs1-content'})
                keywords = paper.find('ul',class_="c-article-subject-list")

                if abstract:
                    if abstract.text:
                        i+=1
                        print(i)

                        abstract = re.sub(r'\n', '', abstract.text)

                        headline = paper.find('h1',class_='c-article-title u-h1').text
                        headline = re.sub('\n', '', headline)
                        if keywords:
                            keywords = keywords.find_all('span', itemprop="about")
                            keywords = ", ".join([keyword.text for keyword in keywords])
                            keywords = re.sub(r'\n', '', keywords)
                        else:
                            keywords=''
                        data.append((headline, abstract, keywords, url+paper_link))
```

```python
df = pd.DataFrame(data, columns=['paper name', 'abstract', 'keywords', 'link'])
```
